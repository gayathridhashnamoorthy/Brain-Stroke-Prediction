# Machine_Learning

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#ENCODING THE CATEGORICAL VARIABLES
for column in data.columns:
    if data.columns.dtype==np.number:
        continue
    data[column]=LabelEncoder().fit_transform(data[column])
```

```python
#SPLIT THE VARIABLE DEPENDENT AND INDEPENDENT
x=data.drop(['stroke'],axis=1)
y=data.stroke
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
y.value_counts()
```

```python
ros=RandomOverSampler(sampling_strategy=1)
x_ros,y_ros=ros.fit_resample(x,y)
ax=y_ros.value_counts().plot.pie(autopct='%.2f')
_=ax.set_title("over_sampling")
```

```python
rus=RandomUnderSampler(sampling_strategy=1)
x_res,y_res=rus.fit_resample(x,y)
ax=y_res.value_counts().plot.pie(autopct='%.2f')
_=ax.set_title("Under_sampling")
```

```python
y_res.value_counts()
```

```python
y_ros.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=.20)
```

```python
x_train_ros,x_test_ros,y_train_ros,y_test_ros=train_test_split(x_ros,y_ros,test_size=.20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.fit(x_train_ros,y_train_ros)
```

```python
pred_ros_lr=lr.predict(x_test_ros)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(pred_ros_lr,y_test_ros)
```

```python
print(classification_report(pred_ros_lr,y_test_ros))
```
